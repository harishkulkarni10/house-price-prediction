###  **Grouping Features by Type**

Before we proceed with encoding and feature engineering, it's important to categorize features based on their nature.

- **Numerical features** can be used directly in regression models, but may require transformation (e.g., log).
- **Categorical features** need to be encoded, using one-hot encoding or label encoding.
- **Ordinal features** represent ordered categories and should be encoded using a ranking map, not dummy variables.
- **Binary features** with only 2 values can be mapped to 0/1 directly.

This grouping helps plan:
- Preprocessing strategies (missing value fills, encoding)
- Feature engineering logic (which variables can be combined)
- Modeling accuracy (e.g., avoiding wrong encodings)

```python
numerical_features = df.select_dtypes(include=["int64", "float64"]).columns.tolist()

categorical_features = df.select_dtypes(include=["object"]).columns.tolist()

binary_features = [col for col in df.columns if df[col].nunique() == 2]

ordinal_features = [
    "ExterQual", "ExterCond", "BsmtQual", "BsmtCond",
    "HeatingQC", "KitchenQual", "FireplaceQu",
    "GarageQual", "GarageCond", "PoolQC"
]
```

```python
binary_features = list(set(binary_features) & set(categorical_features))

ordinal_features = list(set(ordinal_features) & set(categorical_features))

categorical_features = list(set(categorical_features) - set(ordinal_features) - set(binary_features))
```

```python
print("Numerical features:", len(numerical_features))
print("Categorical features:", len(categorical_features))
print("Ordinal features:", len(ordinal_features))
print("Binary features:", len(binary_features))
```

### ** Target Variable (`SalePrice`) Distribution**

Understanding the distribution of our target is essential for regression modeling.

- `SalePrice` is **right-skewed** - there are a few very expensive properties.
- Linear models (like Ridge, Lasso) assume the target is normally distributed.
- A **log transformation** (`np.log1p()`) can help reduce skewness and improve model predictions.
- Also helps identify **outliers** (extremely high prices) that might need handling later.

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

##### Original SalePrice distribution

```python
plt.figure(figsize=(12, 6))
sns.histplot(df['SalePrice'], kde=True, color='skyblue')
plt.title('Distribution of SalePrice (Original)', fontsize=14)
plt.xlabel('SalePrice')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
skew_value = df['SalePrice'].skew()
print(f"Skewness of SalePrice: {skew_value:.2f}")
```

##### Log transformed distribution

```python
plt.figure(figsize=(12, 6))
sns.histplot(np.log1p(df['SalePrice']), kde=True, color='salmon')
plt.title('Distribution of SalePrice (Log Transformed)', fontsize=14)
plt.xlabel('Log(SalePrice)')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

###  **Correlation Matrix of Numerical Features**

This heatmap shows how strongly each numerical feature correlates with others, including `SalePrice`.

Key benefits:
- Helps identify **strong predictors** of house price.
- Reveals **redundant features** (highly correlated with each other).
- Informs **feature selection** and **feature engineering** decisions later.

```python
numeric_df = df.select_dtypes(include=["int64", "float64"])

corr_matrix = numeric_df.corr()

# Plot the heatmap
plt.figure(figsize=(16, 12))
sns.heatmap(corr_matrix, cmap="coolwarm", center=0, annot=False, fmt=".2f")
plt.title("Correlation Matrix of Numerical Features", fontsize=16)
plt.show()
```

```python
cor_target = corr_matrix["SalePrice"].sort_values(ascending=False)

top_corr_features = cor_target[1:11].index.tolist()

top_corr_matrix = df[top_corr_features + ['SalePrice']].corr()

plt.figure(figsize=(10, 8))
sns.heatmap(top_corr_matrix, annot=True, cmap="coolwarm", fmt=".2f", center=0)
plt.title("Top 10 Correlated Features with SalePrice", fontsize=14)
plt.show()
```

```python
def detect_outliers_iqr(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = df[(df[column] < lower_bound) | (df[column] > upper_bound)]
    return outliers

# Detect columns that have more than a given % of outliers
def get_columns_with_outliers(df, threshold=0.01):
    outlier_cols = []
    numeric_cols = df.select_dtypes(include=['int64', 'float64']).columns

    for col in numeric_cols:
        outliers = detect_outliers_iqr(df, col)
        if len(outliers) / len(df) > threshold:
            outlier_cols.append(col)

    return outlier_cols

outlier_columns = get_columns_with_outliers(df)
print("Columns with potential outliers:", outlier_columns)
```

```python
for col in outlier_columns:
    plt.figure(figsize=(5, 2))
    sns.boxplot(x=df[col])
    plt.title(f'Boxplot of {col}')
    plt.grid(True)
    plt.show()
```

##### **Outlier Detection Summary**

Using the IQR (Interquartile Range) method, the following features were found to contain significant outliers:

- `Gr Liv Area`
- `Lot Area`
- `Garage Area`

- These will be considered for capping, transformation, or removal during preprocessing, especially for linear models which are sensitive to outliers. 
- Tree-based models can handle them better, so no strict treatment may be needed there.
